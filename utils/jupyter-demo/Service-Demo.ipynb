```python
import os
from num2words import num2words
import pandas as pd
upper = int(os.getenv('UPPER', '42'))
df = pd.DataFrame({'nums': range(1, upper + 1), 'words': list(map(lambda n: num2words(n), range(1, upper + 1)))})
df
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([1, upper + 1])

plt.plot(ypoints, linestyle = 'dotted')
plt.show()
```
